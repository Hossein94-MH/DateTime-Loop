```python
## sixth homework
patron = input('Patrons ?')      # Patrons ‌‌Branch
if patron == 'Full' :
    waitestimate = float ( input ('waitEstimate ?') )        # waitEstimate Branch
    if waitestimate > 60 :
        print('no')
    elif 0 <= waitestimate <= 10 :
        print('yes')
    elif 10 < waitestimate <= 30 :
        hungry = input('Hungry ? ')        # Hungry Branch from waitEstimate
        if hungry == 'Yes' :
            alternate = input ('Alternate ?')         # Alternate Branch from Hungry
            if alternate == 'yes' :
                raining = input ('Raining ?')      # Raining Branch from Alternate
                if raining == 'yes' :
                    print('yes')
                elif raining == 'no' :
                    print ('no')
                else :
                    print ('wrong entry')
            elif alternate == 'no' :
                print ('yes')
            else :
                print ('wrong entry')    
        elif hungry == 'No' :
            print('yes')
        else : 
            print('wrong entry')

    elif 30 < waitestimate <= 60 :
        alternate = input ('Alternate ?')        # Alternate Branch from waitEstimate 
        if alternate == 'yes' :
            fri_sat = input ('Fri/Sat ?')        # Fri/Sat Branch from Alternate
            if fri_sat == 'yes' :
                print('Yes')
            elif fri_sat == 'no' :
                print('No')
            else :
                print('wrong entry')
        elif alternate == 'no' :
            reservation = input ('Reservation ?')      # Reservation Branch from Alternate
            if reservation == 'yes':
                print('Yes')
            elif reservation == 'no' :
                 bar = input ('Bar ?')      # Bar Branch from Reservation        
                 if bar == 'yes' :
                     print ('Yes')
                 elif bar == 'no' :
                     print('No')
                 else:
                     print('wrong entry')
            else:
                print('wrong entry')
        else:
            print('wrong entry')
    else:
        print ('wong entry')
elif patron == 'some' :
    print('yes')
elif patron == 'none' :
    print('no') 
else :
    print('wrong entry')
```
